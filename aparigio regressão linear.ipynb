```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn import metrics
from sklearn.datasets import load_boston
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
dataset = load_boston()
data = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

### Questões

```python
# 1 - Mostre os primeiros registros da tabela
data.head()
```

```python
# 2 - Observações(linhas) possui nessa base? Quantas Colunas?
print(str(data.shape[0]) + ' linhas e ' + str(data.shape[1]) + ' colunas')
```

```python
# 3 - Quantas Features Possui?
dataset['feature_names']
```

```python
# 4 - Qual é o campo Target(respostas) dessa base?
dataset['target']
```

```python
# 5 - Usando a biblioteca de visualização seaborn, plote o gráfico que mostra a relação entre as features e responses
data['target'] = dataset.target
sns.pairplot(data, x_vars=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX'], y_vars='target', size=7, aspect=0.7, kind='reg')
sns.pairplot(data, x_vars=['RM', 'AGE', 'DIS', 'RAD', 'TAX'], y_vars='target', size=7, aspect=0.7, kind='reg')
sns.pairplot(data, x_vars=['PTRATIO', 'B', 'LSTAT'], y_vars='target', size=7, aspect=0.7, kind='reg')
```

```python
# 6 - Prepare X e y usando o pandas

feature_cols = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
x = data[feature_cols]

y = data['target']
y = data.target
```

```python
# 7 - Qual o tipo de dados de X e y?
print('Os tipos de dados de X e Y são respectivamente: ' + str(type(x)) +', '+ str(type(y)))
```

```python
# 8 - Sobre o que se trata essa base de dados? Que tipo de informações ela guarda?
```

```python
# 9 - Gere um X de treino e y de treino, X de teste e y da base (Split Train/Test)
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# 10 - Qual o percentual gerado para criar o conjunto de treino e o conjunto de teste?
print('Conjunto de treino {0:.2f}%'.format((x_train.shape[0] / (x_train.shape[0] + x_test.shape[0]) * 100)))
print('Conjunto de teste {0:.2f}%'.format((x_test.shape[0] / (x_train.shape[0] + x_test.shape[0]) * 100)))
```

```python
# 11 - Usando modelo de regressão linear do sklearn, treine o modelo com o X e y de treino
linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
# 12 - Imprima os atributos de "intercept" e "coefficients" que foram gerados
print(linreg.intercept_)
print(linreg.coef_)
```

```python
# 13 - imprima o par "feature names" com os "coefficients"
list(zip(feature_cols, linreg.coef_))
```

```python
# 14 - Faça uma previsão usando o conjunto de X de teste
y_pred = linreg.predict(x_test)
print(y_pred)
```

```python
# 15 - Mostre a margem de erro, usando o método "Root Mean Squared Error (RMSE)"
true = [100, 50, 30, 20]
pred = [90, 50, 50, 30]
print('Margem de erro de: {0:.2f}%'.format(np.sqrt(metrics.mean_squared_error(true, pred))))
```
